```python
import torch
import torchvision
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader, random_split
import matplotlib.pyplot as plt
import pandas as pd
import os

# Check for GPU availability
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
torch.backends.cudnn.benchmark = True  # Optimize for GPU

# Ensure model directory exists
os.makedirs("models", exist_ok=True)

# Transforms for 224x224 (MobileNet and AlexNet)
transform_224 = transforms.Compose([
    transforms.RandomCrop(32, padding=4),
    transforms.RandomHorizontalFlip(),
    transforms.Resize((224, 224)),  # Resize for AlexNet and MobileNet
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010)),
])

# Transforms for 64x64 (CustomCNN and ResNet-18)
transform_64 = transforms.Compose([
    transforms.RandomCrop(32, padding=4),
    transforms.RandomHorizontalFlip(),
    transforms.Resize((64, 64)),  # Smaller size for faster training
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010)),
])

# CIFAR-10 Dataset
train_dataset_224 = datasets.CIFAR10(root='./data', train=True, transform=transform_224, download=True)
test_dataset_224 = datasets.CIFAR10(root='./data', train=False, transform=transform_224, download=True)

train_dataset_64 = datasets.CIFAR10(root='./data', train=True, transform=transform_64, download=True)
test_dataset_64 = datasets.CIFAR10(root='./data', train=False, transform=transform_64, download=True)

# Train/Validation Split
train_size = int(0.8 * len(train_dataset_224))
val_size = len(train_dataset_224) - train_size
train_set_224, val_set_224 = random_split(train_dataset_224, [train_size, val_size])
train_set_64, val_set_64 = random_split(train_dataset_64, [train_size, val_size])

# Data Loaders
batch_size = 128
train_loader_224 = DataLoader(train_set_224, batch_size=batch_size, shuffle=True)
val_loader_224 = DataLoader(val_set_224, batch_size=batch_size, shuffle=False)
test_loader_224 = DataLoader(test_dataset_224, batch_size=batch_size, shuffle=False)

train_loader_64 = DataLoader(train_set_64, batch_size=batch_size, shuffle=True)
val_loader_64 = DataLoader(val_set_64, batch_size=batch_size, shuffle=False)
test_loader_64 = DataLoader(test_dataset_64, batch_size=batch_size, shuffle=False)

# Custom CNN
class SmallCustomCNN(nn.Module):
    def __init__(self):
        super(SmallCustomCNN, self).__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(64 * 16 * 16, 128),  # Adjusted for 64x64 input
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(128, 10)
        )

    def forward(self, x):
        x = self.features(x)
        x = x.view(x.size(0), -1)  # Flatten
        x = self.classifier(x)
        return x


# Load Predefined Models
resnet18 = models.resnet18(pretrained=False)
resnet18.fc = nn.Linear(resnet18.fc.in_features, 10)

alexnet = models.alexnet(pretrained=False)
alexnet.classifier[6] = nn.Linear(alexnet.classifier[6].in_features, 10)

mobilenet = models.mobilenet_v2(pretrained=False)
mobilenet.classifier[1] = nn.Linear(mobilenet.classifier[1].in_features, 10)

# Training Function
def train_model(model, train_loader, val_loader, optimizer, num_epochs=10):
    model.to(device)
    best_val_acc = 0
    train_accs, val_accs, train_losses, val_losses = [], [], [], []

    for epoch in range(num_epochs):
        # Training Phase
        model.train()
        train_loss, correct, total = 0, 0, 0
        for data, target in train_loader:
            data, target = data.to(device), target.to(device)
            optimizer.zero_grad()
            output = model(data)
            loss = nn.functional.cross_entropy(output, target)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()
            _, predicted = torch.max(output, 1)
            total += target.size(0)
            correct += (predicted == target).sum().item()

        train_accs.append(correct / total)
        train_losses.append(train_loss / len(train_loader))

        # Validation Phase
        model.eval()
        val_loss, correct, total = 0, 0, 0
        with torch.no_grad():
            for data, target in val_loader:
                data, target = data.to(device), target.to(device)
                output = model(data)
                loss = nn.functional.cross_entropy(output, target)
                val_loss += loss.item()
                _, predicted = torch.max(output, 1)
                total += target.size(0)
                correct += (predicted == target).sum().item()

        val_acc = correct / total
        val_accs.append(val_acc)
        val_losses.append(val_loss / len(val_loader))

        # Save Best Model
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            torch.save(model.state_dict(), f"models/best_model_{model.__class__.__name__}.pth")

        print(f"Epoch {epoch+1}/{num_epochs}, Train Acc: {train_accs[-1]:.4f}, Val Acc: {val_accs[-1]:.4f}")

    return train_accs, val_accs, train_losses, val_losses, best_val_acc

# Evaluation Function
def evaluate_model(model, test_loader):
    model.to(device)
    model.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for data, target in test_loader:
            data, target = data.to(device), target.to(device)
            output = model(data)
            _, predicted = torch.max(output, 1)
            total += target.size(0)
            correct += (predicted == target).sum().item()
    return correct / total

# Grid Search for Hyperparameter Tuning (4 combinations)
def grid_search(model, train_loader, val_loader, hyperparam_grid, num_epochs=10):
    best_hyperparams = None
    best_val_acc = 0

    for params in hyperparam_grid:
        print(f"Testing with hyperparameters: {params}")
        optimizer = params['optimizer'](model.parameters(), lr=params['lr'], weight_decay=params['weight_decay'])
        train_accs, val_accs, train_losses, val_losses, val_acc = train_model(model, train_loader, val_loader, optimizer, num_epochs)
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            best_hyperparams = params
        plot_curves(train_accs, val_accs, train_losses, val_losses, model.__class__.__name__)

    return best_hyperparams, best_val_acc

# Plot Training and Validation Curves
def plot_curves(train_accs, val_accs, train_losses, val_losses, model_name):
    plt.figure()
    plt.plot(train_accs, label="Train Accuracy")
    plt.plot(val_accs, label="Validation Accuracy")
    plt.title(f"Accuracy Curves for {model_name}")
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.show()

    plt.figure()
    plt.plot(train_losses, label="Train Loss")
    plt.plot(val_losses, label="Validation Loss")
    plt.title(f"Loss Curves for {model_name}")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    plt.show()

# Hyperparameter Grid (for Adam and SGD optimizers with different learning rates and weight decays)
hyperparam_grid_adam = [
    {'optimizer': optim.Adam, 'lr': 0.001, 'weight_decay': 0.0},
    {'optimizer': optim.Adam, 'lr': 0.001, 'weight_decay': 0.0001},
    {'optimizer': optim.Adam, 'lr': 0.01, 'weight_decay': 0.0},
    {'optimizer': optim.Adam, 'lr': 0.01, 'weight_decay': 0.0001}
]

hyperparam_grid_sgd = [
    {'optimizer': optim.SGD, 'lr': 0.01, 'weight_decay': 0.0},
    {'optimizer': optim.SGD, 'lr': 0.01, 'weight_decay': 0.0001},
    {'optimizer': optim.SGD, 'lr': 0.001, 'weight_decay': 0.0},
    {'optimizer': optim.SGD, 'lr': 0.001, 'weight_decay': 0.0001}
]

# Train and Evaluate Models
models_224 = [alexnet, mobilenet]
models_64 = [SmallCustomCNN(), resnet18]
results = []


# For models working with 64x64 input (SGD optimizer)
for model in models_64:
    print(f"Starting grid search for {model.__class__.__name__} (64x64 input) with SGD optimizer")
    best_hyperparams, best_val_acc = grid_search(model, train_loader_64, val_loader_64, hyperparam_grid_sgd)
    model.load_state_dict(torch.load(f"models/best_model_{model.__class__.__name__}.pth"))
    test_acc = evaluate_model(model, test_loader_64)
    results.append({
        'Model': model.__class__.__name__,
        'Test Accuracy': test_acc,
        'Best Hyperparameters': best_hyperparams,
        'Validation Accuracy': best_val_acc,
    })

# For models requiring 224x224 input (Adam optimizer)
for model in models_224:
    print(f"Starting grid search for {model.__class__.__name__} (224x224 input) with Adam optimizer")
    best_hyperparams, best_val_acc = grid_search(model, train_loader_224, val_loader_224, hyperparam_grid_adam)
    model.load_state_dict(torch.load(f"models/best_model_{model.__class__.__name__}.pth"))
    test_acc = evaluate_model(model, test_loader_224)
    results.append({
        'Model': model.__class__.__name__,
        'Test Accuracy': test_acc,
        'Best Hyperparameters': best_hyperparams,
        'Validation Accuracy': best_val_acc,
    })

# Results Table and Bar Chart
df_results = pd.DataFrame(results)
print(df_results)

df_results.plot(x="Model", y="Test Accuracy", kind="bar", legend=False, title="Test Accuracy Comparison")
plt.ylabel("Test Accuracy")
plt.show()
```

```python
# Function to format hyperparameters for better readability
def format_hyperparams(hyperparams):
    return f"LR: {hyperparams['lr']}, WD: {hyperparams['weight_decay']}"

# Apply the formatting function to the Best Hyperparameters column
df_results['Best Hyperparameters'] = df_results['Best Hyperparameters'].apply(lambda x: format_hyperparams(x))

# Display the results table in a more readable format
print("\n--- Model Comparison Results ---")
print(df_results.to_string(index=False))  # Print the table without the index

# Print details for all models
print("\n--- Model Details ---")
for _, row in df_results.iterrows():
    print(f"\nModel: {row['Model']}")
    print(f"Test Accuracy: {row['Test Accuracy']:.4f}")
    print(f"Best Hyperparameters: {row['Best Hyperparameters']}")
    print(f"Validation Accuracy: {row['Validation Accuracy']:.4f}")

# Plot Test Accuracy Comparison as a Bar Chart
plt.figure(figsize=(10, 6))
df_results.plot(x="Model", y="Test Accuracy", kind="bar", legend=False, title="Test Accuracy Comparison")
plt.ylabel("Test Accuracy")
plt.xlabel("Model")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Adjust layout for better fit
plt.show()
```
